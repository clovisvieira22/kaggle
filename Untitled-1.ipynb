```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
br_payment_methods = pd.read_csv('brazilian_payment_methods.csv')
```

```python
br_payment_methods['YearMonth'].max()
```

```python
br_payment_methods['YearMonth'] = pd.to_datetime(br_payment_methods['YearMonth'], format='%Y%m')

list_thousands = [1000,2000,3000,4000,5000]
plt.figure(figsize=(14, 7))
y = list_thousands
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['quantityPix'], label='PIX')
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['quantityTED'], label='TED')
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['quantityTEC'], label='TEC')
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['quantityBankCheck'], label='Bank Check')
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['quantityBrazilianBoletoPayment'], label='Brazilian Boleto')
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['quantityDOC'], label='DOC')
plt.xlabel('YearMonth')
plt.ylabel('Number of transactions (in thousands)')
plt.title('Number of transactions across different payment methods over time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['valuePix'], label='PIX')
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['valueTED'], label='TED')
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['valueTEC'], label='TEC')
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['valueBankCheck'], label='Bank Check')
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['valueBrazilianBoletoPayment'], label='Brazilian Boleto')
plt.plot(br_payment_methods['YearMonth'], br_payment_methods['valueDOC'], label='DOC')
plt.xlabel('YearMonth')
plt.ylabel('Value (in millions)')
plt.title('Value of Transactions for Different Payment Methods Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Correlation matrix
correlation_matrix = br_payment_methods[['quantityPix', 'valuePix', 'quantityTED', 'valueTED', 'quantityTEC', 'valueTEC', 'quantityBankCheck', 'valueBankCheck', 'quantityBrazilianBoletoPayment', 'valueBrazilianBoletoPayment', 'quantityDOC', 'valueDOC']].corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Transaction Quantities and Values')
plt.show()
```

```python
quantity_columns = [col for col in br_payment_methods.columns if col.startswith('quantity')]
br_payment_quantities = br_payment_methods[['YearMonth']].copy()
br_payment_quantities = br_payment_methods[quantity_columns]

monthly_growth_rate = br_payment_quantities.groupby('YearMonth').mean().pct_change() * 100

# Verificar os dados aps o clculo da taxa de crescimento
print(monthly_growth_rate.head())

# Plotar a taxa de crescimento mensal
monthly_growth_rate.plot(kind='line', figsize=(14, 7))
plt.title('Monthly Growth Rate of Different Payment Methods')
plt.ylabel('Growth Rate (%)')
plt.xticks(rotation=0)
plt.show()
```

```python
br_payment_quantities = br_payment_methods[['YearMonth'] + quantity_columns]

# Agrupar por 'YearMonth' e calcular a mdia
monthly_means = br_payment_quantities.groupby('YearMonth').mean()

# Calcular a taxa de crescimento percentual
monthly_growth_rate = monthly_means.pct_change() * 100

# Verificar os dados aps o clculo da taxa de crescimento
print(monthly_growth_rate.head())

# Plotar a taxa de crescimento mensal
monthly_growth_rate.plot(kind='bar', figsize=(14, 7))
plt.title('Monthly Growth Rate of Different Payment Methods')
plt.ylabel('Growth Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
