```python
import requests
import pandas as pd

# Funções para obter dados
def get_tfl_bus_arrivals(stop_point_id):
    url = f"https://api.tfl.gov.uk/StopPoint/{stop_point_id}/Arrivals"
    response = requests.get(url)
    return response.json()

def get_tfl_line_status():
    url = "https://api.tfl.gov.uk/Line/Mode/tube/Status"
    response = requests.get(url)
    return response.json()

def get_tfl_disruptions():
    url = "https://api.tfl.gov.uk/Disruptions"
    response = requests.get(url)
    return response.json()

def get_tfl_stop_point(stop_point_id):
    url = f"https://api.tfl.gov.uk/StopPoint/{stop_point_id}"
    response = requests.get(url)
    return response.json()

# Coletando dados de chegadas de ônibus
stop_point_ids = ["490008660N", "490008717N", "490008394E"]
bus_arrivals_data = []
for stop_point_id in stop_point_ids:
    bus_arrivals = get_tfl_bus_arrivals(stop_point_id)
    for arrival in bus_arrivals:
        bus_arrivals_data.append({
            "stop_point_id": stop_point_id,
            "line_name": arrival['lineName'],
            "destination_name": arrival['destinationName'],
            "time_to_station": arrival['timeToStation'] // 60
        })

# Coletando dados de status das linhas do metrô
line_status_data = []
line_status = get_tfl_line_status()
for line in line_status:
    line_status_data.append({
        "line_id": line['id'],
        "line_name": line['name'],
        "status_description": line['lineStatuses'][0]['statusSeverityDescription']
    })

# Coletando dados de interrupções
disruptions_data = []
disruptions = get_tfl_disruptions()
for disruption in disruptions:
    disruptions_data.append({
        "category": disruption['category'],
        "description": disruption['description'],
        "closure_text": disruption['closureText']
    })

# Convertendo para DataFrames
bus_arrivals_df = pd.DataFrame(bus_arrivals_data)
line_status_df = pd.DataFrame(line_status_data)
disruptions_df = pd.DataFrame(disruptions_data)

# Salvando os DataFrames em arquivos CSV
bus_arrivals_df.to_csv('bus_arrivals.csv', index=False)
line_status_df.to_csv('line_status.csv', index=False)
disruptions_df.to_csv('disruptions.csv', index=False)

print("Dados coletados e salvos com sucesso.")
```
